```python

icd_desc_counts = df['PRIMARY_ICD_DESC'].value_counts()
icd_desc_counts_filtered = icd_desc_counts[icd_desc_counts > 10]

plt.figure(figsize=(10, 6))
icd_desc_counts_filtered.plot(kind='bar', color='skyblue')
plt.title('Frequency of Primary ICD Descriptions (Counts > 10)') 
plt.xlabel('Primary ICD Descriptions')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout() 
plt.show()
```

```python
icd_desc_counts_big = df_big['PRIMARY_ICD_DESC'].value_counts()
icd_desc_counts_filtered_big = icd_desc_counts_big[icd_desc_counts_big > 30]

plt.figure(figsize=(10, 6))
icd_desc_counts_filtered_big.plot(kind='bar', color='skyblue')
plt.title('Frequency of Primary ICD Descriptions (Counts > 10)') 
plt.xlabel('Primary ICD Descriptions')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout() 
plt.show()
```

```python
# Filter counts greater than 10
icd_desc_counts_filtered_big = icd_desc_counts_big[icd_desc_counts_big > 25]

# Filter the DataFrame to include only rows where the count is greater than 10
filtered_df_big = df_big[df_big['PRIMARY_ICD_DESC'].isin(icd_desc_counts_filtered_big.index)]

# Plot the stacked bar chart
plt.figure(figsize=(12, 8))  # Set the figure size
sns.countplot(data=filtered_df_big, x='PRIMARY_ICD_DESC', hue='GENDER', palette='coolwarm', order=icd_desc_counts_filtered_big.index, dodge=True)
plt.title('Frequency of Primary ICD Descriptions (Counts > 25) by Gender')  # Set the title of the plot
plt.xlabel('Primary ICD Descriptions')  # Set the label for the x-axis
plt.ylabel('Frequency')  # Set the label for the y-axis
plt.xticks(rotation=45, ha='right') 
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

```python
sec_icd_desc_counts = df['SECONDARY_ICD_DESC'].value_counts()
sec_icd_desc_counts_filtered = sec_icd_desc_counts[sec_icd_desc_counts > 10]

# Filter the DataFrame to include only rows where the count is greater than 10
sec_filtered_df = df[df['SECONDARY_ICD_DESC'].isin(sec_icd_desc_counts_filtered.index)]

# Plot the stacked bar chart
plt.figure(figsize=(12, 8))  # Set the figure size
sns.countplot(data=sec_filtered_df, x='SECONDARY_ICD_DESC', hue='GENDER', palette='coolwarm', order=sec_icd_desc_counts_filtered.index, dodge=True)
plt.title('Frequency of SECONDARY ICD Descriptions (Counts > 10) by Gender')  # Set the title of the plot
plt.xlabel('Secondary ICD Descriptions')  # Set the label for the x-axis
plt.ylabel('Frequency')  # Set the label for the y-axis
plt.xticks(rotation=45, ha='right') 
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```
